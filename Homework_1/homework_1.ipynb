## №2
Визуализируйте распределение количественных переменных с помощью гистограмм и box plot.

```python
# Получаем все столбцы, кроме последнего, который является категориальным
columns_to_plot = df_red.columns[:-1]

fig, axes = plt.subplots(nrows=(len(columns_to_plot)+1)//3, ncols=3, figsize=(15, 5*((len(columns_to_plot)+1)//3)))
fig.suptitle('Гистограммы и плотности распределения для признаков красного вина', fontsize=16)

axes = axes.flatten()

for i, column in enumerate(columns_to_plot):
    # Строим гистограмму
    df_red[column].hist(ax=axes[i], bins=20, density=True, alpha=0.7, color='maroon')
    
    # Добавляем линию плотности
    sns.kdeplot(data=df_red[column], ax=axes[i], color='black', lw=2)
    
    axes[i].set_title(column)
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Плотность')

# Удаляем лишние подграфики в последней тройки
for i in range(len(columns_to_plot), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.subplots_adjust(top=0.95)
plt.show()
```

## №3
Постройте столбчатые диаграммы для анализа распределения категориальных переменных

```python
red_counts = df_red['quality'].value_counts().sort_index()
white_counts = df_white['quality'].value_counts().sort_index()
red_pct = red_counts / red_counts.sum() * 100
white_pct = white_counts / white_counts.sum() * 100
```

```python
red_counts = df_red['quality'].value_counts().sort_index()
white_counts = df_white['quality'].value_counts().sort_index()
red_pct = red_counts / red_counts.sum() * 100
white_pct = white_counts / white_counts.sum() * 100

plt.figure(figsize=(8, 6))
sns.barplot(x=red_counts.index, y=red_pct.values, color='maroon', alpha=0.5, label='Red Wine')
sns.barplot(x=white_counts.index, y=white_pct.values, color='#d8a903', alpha=0.5, label='White Wine')

plt.title('Сравнение распределения по качеству')
plt.xlabel('Quality')
plt.ylabel('Частота')
plt.legend()
plt.show()
```
